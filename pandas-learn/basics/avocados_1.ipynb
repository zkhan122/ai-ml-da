```python
# go by specific constraint e.g. records with Albany region

# define new dataframe
albany_df = df[ df["region"] == "Albany"] # select * from df where region == "Albany"
albany_df.head()
```

```python
# let's check the index
albany_df.index

# we can see from output, there is a index, well we need to ask ourselves, 
# how is the data related? well, we can see that this dataset represents
# the sale of AVOCADOS OVER TIME hence it is being indexed by DATE (as primary key)
```

```python
# if we wanted to index the data by a column e.g. Date( which it is secondarily indexed by already)
# therefore .set_index() returns a data fram 
albany_df.set_index("Date")
```

```python
# now lets check the index
albany_df.head()

# as we can see, we lost our index (Pandas limitations), it actually returns a new dataframe
```

```python
# to fix this issue 
# first fix - reassign
albany_df = albany_df.set_index("Date")
# nothing gets printed out, it is just reassigned
```
