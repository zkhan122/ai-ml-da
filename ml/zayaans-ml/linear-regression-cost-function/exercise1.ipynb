## Linear Regression with One Variable (Uni-Variate)

Implementing linear regression with one variable to predict profits for a food truck. I have data for profits and populations from the cities. I would like to use this data to help you select which city to expand to next. 

Given `Data/ex1data1.txt` it contains dataset for our linear regression problem. The first column is population of city (per 1000s) and 2nd column is profit of food truck in that city (in $10,000s). Negative value indicates a loss

We must load each columns into variable `x` and `y`.

```python
data = np.loadtxt(os.path.join("Data", "ex1data1.txt"), delimiter=",")
X, y = data[:, 0], data[:, 1] # X and y both have same length 

for i in range(len(X)):
    print(X[i], y[i])
```

```python
print("No. of cities:", X.size, "\nNo. of profits/losses:", y.size)
```

```python
# now we must plot the data
def plotData(populationData, ProfitLossData):
    fig = plt.figure()
    plt.plot(X, y, "ro", ms=10, mec="black")
    plt.xlabel("Population per city (1000s)")
    plt.ylabel("Values ($10000)")
    """ 
    ro represents the circles
        ms is marker size
        mec is marker edge color (border color) """

plotData(X, y)
```

```python
# now we must plot the data
def plotDataWithLineOfBestFit(populationData, ProfitLossData):
    # peform linear regression
    # f(x) = Ax+B
    a, b = np.polyfit(X, y, 1) # a is gradient and b is intercept
    lineOfBestFit = a*X + b
    plt.plot(X, y, "ro", ms=10, mec="black")
    plt.plot(X, lineOfBestFit, label="Line of Best Fit")
    plt.xlabel("Population per city (1000s)")
    plt.ylabel("Value per city ($10000s)")
    plt.legend()
    plt.show()
    
```

```python
plotDataWithLineOfBestFit(X, y)
```
